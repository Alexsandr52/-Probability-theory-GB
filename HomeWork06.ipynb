<h4>Задание 1</h4>
Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):<br>
    zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],<br>
    ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].<br>
    Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy
    Полученные значения должны быть равны.<br>
    Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков,
    а затем с использованием функций из библиотек numpy и pandas.


cov $= M(X \cdot Y) - M(x) \cdot M(Y)$

```python
zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]

zp = np.array(zp)
ks = np.array(ks)

cov_matrix = np.cov(zp,ks, ddof=0)
cov_value = (ks*zp).mean() - zp.mean() * ks.mean()

print(cov_matrix)
print(f'Значение полученное с помощью элементарных действий - {cov_value}')

```

### $r_{xy} = \frac{cov_{xy}}{std(x)\cdot std(y)}$

```python
std_zp = zp.std(ddof=0)
std_ks = ks.std(ddof=0)
print(f'корреляция Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков - {cov_value / (std_zp * std_ks)}')
np.corrcoef(zp, ks)
```

<h3>Задание 2</h3>
Измерены значения IQ выборки студентов,<br>
    обучающихся в местных технических вузах:<br>
    131, 125, 115, 122, 131, 115, 107, 99, 125, 111.<br>
    Известно, что в генеральной совокупности IQ распределен нормально.<br>
    Найдите доверительный интервал для математического ожидания с надежностью 0.95.<br>

M $= 118.1$

std $= 10.5456$

a $= 5\% == 2.5\% \cdot 2$

9 степеней свобод и a $= 5\%$

<img src = 'https://www.matematicus.ru/wp-content/uploads/2019/02/Znacheniya-kriteriya-Styudenta-pri-urovne-znachimosti-0.1-0.05-0.01.png'>


```python
data = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])

t = 2.26
M = data.mean()
std = data.std(ddof=1)
n = 10

confidence_interval = [M - (t * std) / (n ** 0.5), M + (t * std) / (n ** 0.5)]
f'Интервал {confidence_interval}'
```

Известно, что рост футболистов в сборной распределен нормально<br>

Тоже что и во втором задании только здесь используем Z критерий так как есть std генеральной совокупности.

std = $\sqrt{D} = \sqrt{25} = 5$

M $= 174.2$

a $= 5\%$

n $= 27$

<img src='https://storage.yandexcloud.net/wr4img/164966_81__176.png'>

```python
Z = 1.96
M = 174.2
std = 5
n = 27

confidence_interval = [M - (Z * std) / (n ** 0.5), M + (Z * std) / (n ** 0.5)]
f'Интервал {confidence_interval}'
```
