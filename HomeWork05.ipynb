<strong>Известно, что генеральная совокупность распределена нормально
со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,
если выборочная средняя M = 80, а объем выборки n = 256.</strong>

std = $16$

M = $80$

n = $256$

a = $5%$ == $2.5$ с одной стороны

$a = m + Z_1 \cdot \frac{std}{\sqrt{n}}$

$80 - 1.96 \cdot \frac{16}{\sqrt{256}}$

$b = m + Z_2 \cdot \frac{std}{\sqrt{n}}$

$80 + 1.96 \cdot \frac{16}{\sqrt{256}}$

```python
se = 16 / 256**(0.5)
a = 80 - 1.96 * se
b = 80 + 1.96 * se
[a, b]
```

<strong>Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.</strong>

Из условия задачи

$x = 17.5$

$std = \sqrt{D} = 2$

$n = 100$

$a = 0.5$

$H_0$ - Значемых различий нет

$H_0$ - Значемые различия есть

Используем Z критерий так как нам известна std

$Z = 1.65$ По таблице выше (1 задание)

$se = \frac{std}{\sqrt{n}} = \frac{2}{\sqrt{100}} = 0.2$

зная что se 0.2 

Z * se  ?  17 - 17.5

```python
Z = 1.65
se = 0.2
print((Z * se) > (17.5 - 17))
```

Вывод верна теория $H_1$

<strong>Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)</strong>

Из условия задачи

$x = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]$

$std_0 = 4.453$

$m = 198.5$

$n = 10$

$a = 0.01$

$H_0$ - Утверждение верно

$H_0$ - Утверждение неверно

Пользуемся критерием T так как нет std, a для каждой стороны 0.005

```python
data = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
print(f'M = {data.mean()}, std_0 = {data.std(ddof=1)}')
```

```python
M = 200
m = 198.5
t = 3.2498
se = data.std(ddof=1) / 10**(0.5)
print((m-M)+ t*se < t)
print(-(m-M) - t*se > -t)

```

Вывод верна теория $H_0$
