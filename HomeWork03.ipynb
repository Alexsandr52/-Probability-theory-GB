<strong>1 Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.</strong>

##### Cреднее арифметическое
##### $M(X) = \frac{1}{n} \sum\limits_{i=1}^{n} x_i = 65.3$

##### Cреднее квадратичное отклонение
##### $\sigma = \sqrt{\frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}} = 30.82$

##### Cмещенная оценка дисперсии
##### $\sigma^2 = \frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n} = 950.11$

##### Не смещенная оценка дисперсии
##### $\sigma^2_{нес} = \frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n - 1} = 1000.11$


```python
from cmath import sqrt


salary_list = pd.Series([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])

my_mean = salary_list.sum() / salary_list.count()
print(f'Среднее арифметическое - {my_mean}, среднее арифметическое numpy - {np.mean(salary_list)}')

my_std = np.sqrt(((salary_list - my_mean)**2).sum() / salary_list.count())
print(f'Среднее квадратичное отклонение - {my_std}, среднее квадратичное отклонение numpy - {np.std(salary_list)}')

my_var_ddof0 = my_std**2                          #((salary_list - my_mean)**2).sum() / salary_list.count()
print(f'Смещенная оценка дисперсии {my_var_ddof0}, смещенная оценка дисперсии numpy {np.var(salary_list)}')

my_var_ddof1 = ((salary_list - my_mean)**2).sum() / (salary_list.count() - 1)
print(f'Не смещенная оценка дисперсии {my_var_ddof1}, смещенная оценка дисперсии pandas {salary_list.var(ddof = 1)}')

```

```python
from math import factorial
# сочетание
def C(k, n):
    return factorial(n) / (factorial(k) * factorial(n - k))

# Размещение
def A(k, n):
    return factorial(n) / factorial(n - k)

# Перестановка
def P(n):
    return factorial(n)
```

<strong>3 На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. #### Воспользуемся формулой Байеса $P(Стрелял спортсмен n| было поподание) = \frac{\frac{1}{3} * вероятность поподания n спортсмена}{полная вероятность(P_a)}$

$P_a = \frac{1}{3} \cdot 0.9 + \frac{1}{3} \cdot 0.8 + \frac{1}{3} \cdot 0.6 = 0.7(6)$

a) 0.39 = 39%

б) 0.34 = 34%

в) 0.26 = 26%

```python
pa = 1/3 * 0.9 + 1/3 * 0.8 +1/3 * 0.6
p1 = 1/3 * 0.9 / pa
p2 = 1/3 * 0.8 / pa
p3 = 1/3 * 0.6 / pa

print(p1, p2, p3)
```
