Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?</li><br>

```python
from math import factorial, exp
# сочетание
def C(k, n):
    return factorial(n) / (factorial(k) * factorial(n - k))

# формула Бернулли
def P(n, x, p):
    return C(x, n) * (p ** x) * (1 - p) ** (n-x) 

def Puasson(l, m):
    return l ** m / factorial(m) * exp(-l)
```

<strong>Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.</strong>

Количество испытаний не большое и вероятность не слишком мала, воспользуемся формулой Бернули

##### $P_{100}(85) = C^{85}_{100} 0.8^{85} 0.8^{100-85} ≈ 0.048 ≈ 4.8\%$

```python
P(100, 85, 0.8)
```

<strong>Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?</strong>

Большое число испыаний и маленькая вероятность, используем распределение Пуассона

##### $P_m ≈ \frac{\lambda^m}{m!}e^{-\lambda}$


##### $1) P_0 ≈ \frac{\lambda^0}{0!}e^{-5000*0.0004} ≈ 0.13 ≈ 13\%$

##### $2) P_2 ≈ \frac{(5000*0.0004)^2}{2!}e^{-5000*0.0004} ≈ 0.27 ≈ 27\%$

```python
l = 5000 * 0.0004
Puasson(l, 0)
```

```python
Puasson(l, 2)
```

<strong>Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?</strong>

Количество испытаний не большое и вероятность не слишком мала, воспользуемся формулой Бернули

##### $P_{144}(70) = C^{70}_{144} 0.5^{70} 0.5^{144-70} ≈ 0.062 ≈ 6.2\%$

```python
P(144, 70, 0.5)
```
