```python
from math import factorial
# сочетание
def C(k, n):
    return factorial(n) / (factorial(k) * factorial(n - k))

# Размещение
def A(k, n):
    return factorial(n) / factorial(n - k)

# Перестановка
def P(n):
    return factorial(n)
```

<strong>Из колоды в 52 карты извлекаются случайным образом 4 карты. </strong>
<ul>
    <li>a) Найти вероятность того, что все карты – крести. </li>
    <li>б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.</li>
</ul>

В колоде 52 карты => 13 карт каждой масти, для того чтобы найти все возможные варианты выбрать 4 карты крести из 13 возможных используем сочетание, так как порядок карт не важен.

##### $C^{4}_{13} = \frac{13!}{4!(13-4)!} = 715$
Получается что $715$ это количество благоприятных исходов.

Теперь найдем все варианты выбрать 4 карты из колоды.
##### $C^{4}_{52} = \frac{52!}{4!(52-4)!} = 270725$
Разделим количество благоприятных исходов на количество всех возможных исходов
##### $P = \frac{C^{4}_{13}} {C^{4}_{52}} = \frac{715}{270725} = 0.0026410564225690276$

Примерно ровно $0.26 \% $


```python
print(C(4, 13))
print(C(4, 52))
print(715/270725)
```


В колоде 4 туза. Значит что без тузов в колоде 48 карт. Найдем все возможности выбрать из 48 карт 4 карты 

##### $C^4_{48} = \frac{48!}{4!(48-4)!} = 194580$
В данном случае будем считать это как количество благоприятных исходов

Теперь найдем все возможные исходы

##### $C^4_{52} = \frac{52!}{4!(52-4)!} = 270725$

Найдем вероятность что среди этих 4 карт не будет ни одного туза

##### $P_1 = \frac{C^4_{48} }{C^4_{52} } = 0.7187367254594146$

Найдем вероятность что среди этих 4 карт не будет хотя бы одного туз

##### $P_2 = 1 - P_1 = 1 - \frac{C^4_{48} }{C^4_{52} } = 1 - 0.718 = 0.282$



```python
print(C(4, 48))
print(C(4, 52))
print(C(4, 48) / C(4, 52))
1 - C(4, 48) / C(4, 52)
```

<strong>В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?</strong>

Найдем все возможные варианты выбрать 2 билета из 100, причем порядок в данном случае важен. Используем для этого размещение.

##### $A^2_{100} = \frac{100!}{2!(100-2)!} = 9900$

Найдем вероятность разделив количество выбранных билетов на количество выбрать 2 билета из 100

##### $P = \frac{2}{A^2_{100}} = \frac{2}{9900} = 0.000(20) = 0.02%$




```python
print(A(2,100))
2/A(2,100) * 100
```
